# Introduction

 The United Nations Food and Agriculture Organization's publication, “The State of Food Security and Nutrition in the World 2022,” looks into issues of global hunger and food insecurity. While the report focuses on these problems globally, these issues are more persistent in the United States than the report might lead one to conclude.

Using data from the Economic Research Service of the US Department of Agriculture, this assignment will look into the state of food insecurity and it's correlation with poverty levels in the US. Also, we will explore how food insecurity has impacted different households.

# Gathering and Cleaning Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
poverty = pd.read_csv("https://raw.githubusercontent.com/LeJQC/MSDS/main/DATA%20608/Story%206/Food%20Security%20all%20households.csv")
```

```python
poverty = poverty[poverty['Category'] == 'Household income-to-poverty ratio']
poverty = poverty[poverty['Subcategory'] != 'Income unknown']
```

```python
poverty.head()
```

```python
poverty.describe()
```

```python
children = pd.read_csv('https://raw.githubusercontent.com/LeJQC/MSDS/main/DATA%20608/Story%206/Food%20security%20Household%20with%20children.csv')
```

```python
children = children[children['Category'] == 'Household composition']
children = children[children['Year'] >= 2021].sort_values(['Subcategory','Food-insecure households-percent'])
children.head()
```

```python
children
```

```python
states = pd.read_csv('https://raw.githubusercontent.com/LeJQC/MSDS/main/DATA%20608/Story%206/Food%20Security%20by%20state.csv')
```

```python
states =states[states['Year'] == '2020?2022']
states = states[1:]
states = states.sort_values('Food insecurity prevalence',ascending=False)
```

# Data Visualizations

```python
grouped_data = poverty.groupby('Subcategory')

desired_order = ['Under 1.00', 'Under 1.30', 'Under 1.85', '1.85 and over']

fig, ax1 = plt.subplots(figsize=(11, 6))
for i, subcategory in enumerate(desired_order):
    if subcategory in grouped_data.groups:
        group = grouped_data.get_group(subcategory)
        ax1.plot(group['Year'], group['Food insecure-percent'], label=subcategory,
                 color=sns.color_palette("Blues", 4)[::-1][i])
        ax1.scatter(2001, group[group['Year'] == 2001]['Food insecure-percent'], color='darkblue')
        ax1.scatter(2022, group[group['Year'] == 2022]['Food insecure-percent'], color='darkblue')
        ax1.text(2001, group[group['Year'] == 2001]['Food insecure-percent']+1, str(group[group['Year'] == 2001]['Food insecure-percent'].values[0]), color='red',ha='center')
        ax1.text(2022.2, group[group['Year'] == 2022]['Food insecure-percent'], str(group[group['Year'] == 2022]['Food insecure-percent'].values[0]), color='red')

ax1.text(2011, 20, 'The Percentage of Food Insecured Households has INCREASED over the last 20 years', color='red',ha='center',fontsize=12)

ax1.set_xlabel('Year')
ax1.set_ylabel('Food Insecurity by Percentage',color='darkblue')
ax1.set_title('Food Insecurity is More Prevalent in Poorer Households', fontsize=14,color='red')

ax1.legend(loc='upper left', bbox_to_anchor=(1, 0.7), title="Household Income to Poverty Ratio")


plt.show()
```

```python
category_order = ['Married-couple families', 'With children < 6 years', 'Male head, no spouse', 'Other household with child', 'Female head, no spouse']

palette = {2021: "lightgreen", 2022: "darkgreen"}

fig, ax2 = plt.subplots(figsize=(13, 7))

bars = sns.barplot(x='Subcategory', y='Food-insecure households-percent', hue='Year', data=children, ax=ax2, palette=palette, order=category_order)

for bar in bars.patches:
    bar_height = bar.get_height()
    if bar_height > 0:
        bars.annotate(format(bar_height, '.2f'),
                      (bar.get_x() + bar.get_width() / 2,
                       bar_height), ha='center', va='center',
                      size=10, xytext=(0, 8),
                      textcoords='offset points', color = 'red')

ax2.set_xlabel('Subcategory',fontsize=12)
ax2.set_ylabel('Food Insecured Households (%)', color='darkgreen')
ax2.set_title('Households with Children Saw an INCREASE in Food Insecurity in 2022')

ax2.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.show()
```

```python
states = states.sort_values('Food insecurity prevalence',ascending=False)

palette = sns.light_palette("green", 51)
palette.reverse()

fig, ax3 = plt.subplots(figsize=(11, 6))

sns.barplot(x='State', y='Food insecurity prevalence', data=states, palette=palette, legend=False)
plt.xticks(rotation=45)
ax3.set_xlabel('State')
ax3.set_ylabel('Food-Insecure Households (%)', color='darkgreen')
ax3.set_title('Food Insecurity By State ')

ax3.axhline(10, color='r', linestyle='--')
#ax3.text(27, 10.8, '10.8% - Average % of Households that are Food insecured in US', color='r', va='bottom')

ax3.text(10, 15, 'In 33 states(66%),\nOne in Ten Households\nAre Food Insecured',color='red',ha='center',fontsize=12)

plt.tight_layout()
plt.show()
```

# Conclusion

Although food security and nutrition may be a problem globally, 13% or 115,750,000 households in the US are still battling food insecurity. This percentage is significantly higher in households with a income to poverty ratio below 1.85. Remarkedly, this issue has not improved in the past 20 years as the percentage of households that are food insecured in impoverished population has risen since 2000. Not only is it a problem in the poorer households, households with a income to poverty ratio over 1.85 have also seen a slight increase in food insecurity.

In the US, two thirds of the country have a food-insecured percentage of 10% or greater. To put in perspective, that is 1 in 10 families in these states that are food insecured.

Looking at the 2 most recent years of food security data, all households with children have seen a remarkable rise in food security. In female lead households with children, one third of this population is food insecured. This is over a 10% increase from the previous year. These findings have serious implications as the high rates of food insecurity in households with children, particularly those led by females, could have long-term impacts on the health and development of these children. According to this article from the American Association of Pediatrics (https://publications.aap.org/pediatrics/article/144/4/e20190397/38475/Food-Insecurity-and-Child-Health?autologincheck=redirected), children in food insecure households had higher rates of asthma diagnoses, deprssive symptoms, and emergency room visits. These alarming trends underscore the urgent need for action.
